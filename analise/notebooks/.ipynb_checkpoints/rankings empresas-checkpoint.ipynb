```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lic = pd.read_csv("propostas_licitacao.txt", sep="|", dtype={'cod_unidade_gestora': int,'numero_licitacao': str,'cod_tipo_licitacao': int,
                                                   'qtd_ofertada': int,'unidade_gestora': str, 'tipo_licitacao': str, 
                                                   'cpf_cnpj': str,'nome_proponente': str, 'situacao_proposta': str})
```

```python
#Ranking das empresas que mais venceram em licitações.
vencedoras = lic['nome_proponente'].groupby(lic['situacao_proposta']).get_group('Vencedora')
v = vencedoras.value_counts()[:10]
v.plot(kind='barh', x='nome_proponente').invert_yaxis()
plt.title("Top 10 das empresas que mais venceram licitação na PB")
plt.ylabel('Empresas')
plt.xlabel('Quantidade de licitações')
```

```python
#Ranking das empresas que mais perderam em licitações.
perdedoras = lic['nome_proponente'].groupby(lic['situacao_proposta']).get_group('Perdedora')
p = perdedoras.value_counts()[:10]
p.plot(kind='barh', x='nome_proponente').invert_yaxis()
plt.title("Top 10 das empresas que mais perderam licitação na PB")
plt.ylabel('Empresas')
plt.xlabel('Quantidade de licitações')
```

```python
#Ranking das empresas que mais faturaram em licitações (soma do valor das propostas vencedoras).
mais_fatura = lic[lic['situacao_proposta'].str.contains("Vencedora")]
maisfat = mais_fatura.groupby('nome_proponente')['valor_ofertado'].sum()
mf= maisfat.nlargest(10)
mf.plot(kind='barh', x='nome_proponente').invert_yaxis()
plt.title("Top 10 das empresas que mais faturaram em licitação na PB")
plt.ylabel('Empresas')
plt.xlabel('Quanto faturaram')
```

```python
#Ranking das empresas que mais participaram de licitação 
#(cuidado para contar apenas 1x se o participante tiver mais de uma proposta em uma mesma licitação).
mais_participa = lic.groupby('nome_proponente').situacao_proposta.agg(lambda g: g.eq('Vencedora', 'Perdedora').count())
mp = mais_participa.sort_values(ascending=False)[:10]
mp.plot(kind='barh', x='nome_proponente').invert_yaxis()
plt.title("Top 10 das empresas que mais participaram de licitação na PB")
plt.ylabel('Empresas')
plt.xlabel('Quantidade de participações')
```
