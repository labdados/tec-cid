```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lic = pd.read_csv("propostas_licitacao.txt", sep="|", dtype={'cod_unidade_gestora': int,'numero_licitacao': str,'cod_tipo_licitacao': int,
                                                   'qtd_ofertada': int,'unidade_gestora': str, 'tipo_licitacao': str, 
                                                   'cpf_cnpj': str,'nome_proponente': str, 'situacao_proposta': str})
```

```python
#não soma as vencedoras, só junta elas
m = lic.groupby(['nome_proponente','valor_ofertado'])['situacao_proposta'].apply(lambda x: x[x.str.contains('Vencedora')].sum())
m
```

```python
#soma tudo, as vencedoras e perdedoras
çç = (lic.groupby(['nome_proponente','situacao_proposta'])['valor_ofertado'].sum())
çç
```

```python
çç.nlargest(10)
```
