```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lic = pd.read_csv("licitacao.txt", sep="|", dtype={'cod_unidade_gestora': int,'numero_licitacao': str,'cod_tipo_licitacao': int,
                                                   'qtd_ofertada': int,'unidade_gestora': str, 'tipo_licitacao': str, 
                                                   'cpf_cnpj': str,'nome_proponente': str, 'situacao_proposta': str})
```

```python
lic.dtypes
```

```python
lic.groupby('unidade_gestora').count()
```

```python
licitacao = lic.groupby('unidade_gestora')
```

```python
licitacao.size()
```

```python
lic['unidade_gestora'].value_counts()
```

```python
lic['unidade_gestora'].value_counts()[:10]
```

```python
lic['unidade_gestora'].value_counts()[:10].plot(kind='barh')
plt.title("Top 10 das cidades que mais fazem licitação na PB")
plt.ylabel('Cidades')
plt.xlabel('Quantidade de licitações')
```

```python
lic['nome_proponente'].value_counts()[:10].plot(kind='barh')
plt.title("Top 10 das empresas que mais participaram de licitação na PB")
plt.ylabel('Empresas')
plt.xlabel('Quantidade de licitações')
```
