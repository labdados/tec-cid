```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lic = pd.read_csv("licitacao.txt", sep="|", dtype={'cod_unidade_gestora': int,'numero_licitacao': str,'cod_tipo_licitacao': int,
                                                   'qtd_ofertada': int,'unidade_gestora': str, 'tipo_licitacao': str, 
                                                   'cpf_cnpj': str,'nome_proponente': str, 'situacao_proposta': str})
```

```python
lic.dtypes
```

```python
lic['unidade_gestora'].value_counts()[:10]
```

```python
lic['unidade_gestora'].value_counts()[:10].plot(kind='barh')
plt.title("Top 10 das cidades que mais fazem licitação na PB")
plt.ylabel('Unidades Gestoras')
plt.xlabel('Quantidade de licitações')
```

```python
lic['nome_proponente'].value_counts()[:10].plot(kind='barh')
plt.title("Top 10 das empresas que mais participaram de licitação na PB")
plt.ylabel('Empresas')
plt.xlabel('Quantidade de licitações')
```

```python
lic.nlargest(10, ['valor_ofertado'])
```

```python
gastadeiros = lic.nlargest(10, ['valor_ofertado'])
```

```python
gastadeiros.plot(kind='barh', x='unidade_gestora', y='valor_ofertado')
plt.title("Top 10 das cidades que mais gastam com licitação na PB")
plt.ylabel('Unidades Gestoras')
plt.xlabel('Valor das licitações')
```

```python
gastadeiros.plot(kind='barh', x='nome_proponente', y='valor_ofertado')
plt.title("Top 10 das empresas que mais receberam com licitação na PB")
plt.ylabel('Empresas')
plt.xlabel('Valor recebido')
```
