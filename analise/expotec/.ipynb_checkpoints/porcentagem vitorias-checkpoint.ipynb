# Análise dos maiores e menores vencedores
    * Ranking da porcentagem de vitórias
    * Ranking da porcentagem de derrotas     

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lic = pd.read_csv("propostas_licitacao.txt", sep="|", dtype={'cod_unidade_gestora': int,'numero_licitacao': str,'cod_tipo_licitacao': int,
                                                   'qtd_ofertada': int,'unidade_gestora': str, 'tipo_licitacao': str, 
                                                   'cpf_cnpj': str,'nome_proponente': str, 'situacao_proposta': str})
```

```python
porcentagem = (lic['nome_proponente'].value_counts().rename_axis('nome_proponente').to_frame('situacao_proposta')
                   .join(pd.crosstab(lic['nome_proponente'], lic['situacao_proposta'], normalize=0).mul(100).round(1))
                   .reset_index())
porcentagem
```

```python
propostas = porcentagem[:10]
del propostas['situacao_proposta']
propostas.style.format({
    'Perdedora': '{0:.2f}%'.format,
    'Vencedora': '{0:.2f}%'.format,
    'Pesquisado': '{0:.2f}%'.format,
})
```

```python
from matplotlib.ticker import PercentFormatter
propostas.plot(x="nome_proponente", kind="barh").invert_yaxis()
plt.title("Top 10 da porcentagem de empresas que mais fizeram propostas de licitações na Paraíba")
plt.ylabel('Unidades Gestoras')
plt.xlabel('Quantidade de licitações')
plt.gca().xaxis.set_major_formatter(PercentFormatter(decimals=0.2))
```

```python
ganhou = porcentagem.sort_values(by='Vencedora', ascending=False)[:10]
del ganhou['situacao_proposta']
ganhou.style.format({
    'Perdedora': '{0:.2f}%'.format,
    'Vencedora': '{0:.2f}%'.format,
    'Pesquisado': '{0:.2f}%'.format,
})
```

```python
from matplotlib.ticker import PercentFormatter
ganhou.plot(x="nome_proponente", kind="barh").invert_yaxis()
plt.title("Top 10 da porcentagem de empresas que mais venceram propostas de licitações na Paraíba")
plt.ylabel('Unidades Gestoras')
plt.xlabel('Quantidade de licitações')
plt.gca().xaxis.set_major_formatter(PercentFormatter(decimals=0.2))
```

```python
perdeu = porcentagem.sort_values(by='Perdedora', ascending=False)[:10]
del perdeu['situacao_proposta']
perdeu.style.format({
    'Perdedora': '{0:.2f}%'.format,
    'Vencedora': '{0:.2f}%'.format,
    'Pesquisado': '{0:.2f}%'.format,
})
```

```python
from matplotlib.ticker import PercentFormatter
perdeu.plot(x="nome_proponente", kind="barh").invert_yaxis()
plt.title("Top 10 da porcentagem de empresas que mais perderam propostas de licitações na Paraíba")
plt.ylabel('Unidades Gestoras')
plt.xlabel('Quantidade de licitações')
plt.gca().xaxis.set_major_formatter(PercentFormatter(decimals=0.2))
```
