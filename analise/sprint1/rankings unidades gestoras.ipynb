```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lic = pd.read_csv("licitacao.txt", sep="|", dtype={'cd_ugestora': str,'de_ugestora': str,'nu_licitacao': str,
                                                   'tp_licitacao': float,'de_tipolicitacao': str, 'dt_homologacao': str, 
                                                   'tp_objeto': str,'de_tipoobjeto': str, 'vl_licitacao': float,'de_obs': str}, engine='python', error_bad_lines=False, encoding = 'utf8')
```

```python
#Ranking das unidades gestoras por quantidade de licitações feitas
lic['de_ugestora'].value_counts()[:10]
lic['de_ugestora'].value_counts()[:10].plot(kind='barh').invert_yaxis()
plt.title("Top 10 das Unidades Gestoras que mais fazem licitações")
plt.ylabel('Unidades Gestoras')
plt.xlabel('Quantidade de licitações')
```

```python
#Ranking das unidades gestoras por valor total das licitações
qtd_ugestora = lic.groupby(['de_ugestora'], as_index=False)['vl_Licitacao'].sum() #somando as icitações
qtd_ug = qtd_ugestora.sort_values(by ='vl_Licitacao', ascending=False)[:10]  #colocando em ordem decrescente
qtd_ug.plot(kind='barh', x='de_ugestora', y='vl_Licitacao').invert_yaxis()
plt.title("Top 10")
plt.ylabel('Unidades Gestoras')
plt.xlabel('Soma das Licitações')
```

```python
#Ranking das licitações de maior valor
valor_licitacao = lic.nlargest(10, ['vl_Licitacao'])
valor_licitacao.plot(kind='barh', x='de_ugestora', y='vl_Licitacao').invert_yaxis()
plt.title("Top 10")
plt.ylabel('Unidades Gestoras')
plt.xlabel('Valor das licitações')
```
